```python
# imports :
import cv2
import numpy as np
import face_recognition
```

```python

# load the images (image and test image)
__image__ = face_recognition.load_image_file('data/stilefile.jpg')
__image__ = cv2.cvtColor(__image__, cv2.COLOR_BGR2RGB)

__image_test__ = face_recognition.load_image_file('data/djoms.jpeg')
__image_test__ = cv2.cvtColor(__image_test__, cv2.COLOR_BGR2RGB)
```

```python

# displaying the images (image and test image)
cv2.imshow('Stile', __image__)
cv2.imshow('Stile Test', __image_test__)
cv2.waitKey(0)
```

```python

# find and encode the face
face_region = face_recognition.face_locations(__image__)[0]
print(face_region)
face_encodings = face_recognition.face_encodings(__image__)[0]
# print(face_encodings)
face_region_test = face_recognition.face_locations(__image_test__)[0]
print(face_region_test)
face_encodings_test = face_recognition.face_encodings(__image_test__)[0]
# print(face_encodings_test)
```

```python
cv2.rectangle(__image__,
              (face_region[3], face_region[0]),
              (face_region[1], face_region[2]),
              (255, 0, 0), 2)
cv2.imshow('stile face', __image__)
cv2.waitKey(0)

# we could do the same thing for the test image ...
```

```python

# we use the linear svm to compare faces
result = face_recognition.compare_faces([face_encodings], face_encodings_test)
face_distance = face_recognition.face_distance([face_encodings], face_encodings_test)
print(result)
face_distance = round(face_distance[0], 2);
print(face_distance)

cv2.putText(__image_test__,
            f'{result[0]} {face_distance}',
            (50, 50),
            cv2.FONT_HERSHEY_SIMPLEX,
            1,
            (0, 255, 0), 2)
cv2.imshow('djoms', __image_test__)
cv2.waitKey(0)
```
